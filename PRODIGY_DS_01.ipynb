## Import neccessary Libraries


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Import dataset to the kernel

```python
df=pd.read_csv('C:/Users/manov/Downloads/API_SP.POP.TOTL_DS2_en_csv_v2_31753/population.csv',skiprows=4)
```

```python
df.head(5)
```

```python
df=df.drop("Unnamed: 68",axis=1)
df.head(5)
```

```python

df_melted = pd.melt(df, id_vars=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'], 
                    var_name='Year', value_name='Population')

df_melted['Year'] = pd.to_numeric(df_melted['Year'], errors='coerce')

df_melted.head()
```

```python
df_2023=df_melted[df_melted['Year']==2023]
df_2023
```

## Distribution of Population Across Countries (2023)

```python
plt.figure(figsize=(8, 6))
sns.histplot(df_2023['Population'], bins=20, kde=False)
plt.show()
```

```python
df_2023_top10=df_2023.nsmallest(10,'Population')
df_2023_top10
```

## Plotting top 10 countries population in 2023


```python
plt.figure(figsize=(10,6))
sns.barplot(x="Country Name",y="Population",data=df_2023_top10)
plt.xticks(rotation=45)
plt.xlabel("Country")
plt.ylabel("Population")
plt.title("Top 10 Countries with most population")
plt.show()
```
